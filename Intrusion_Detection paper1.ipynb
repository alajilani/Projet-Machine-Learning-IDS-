### Introduction

L'évolution rapide des technologies a entraîné une augmentation significative des menaces liées à la sécurité informatique. Les attaques par intrusion, qu'elles soient externes ou internes, sont devenues de plus en plus sophistiquées, mettant en danger la confidentialité, l'intégrité et la disponibilité des données. Dans ce contexte, la mise en place d'un Intrusion Detection System (IDS) utilisant des techniques de machine learning devient primordiale pour prévenir, détecter et répondre aux activités malveillantes potentielles. Ce projet consiste donc à appliquer différents algorithmes supervisés et non supervisés de machine learning pour detecter ces intrusions malveillantes.

### Familles de systèmes de détection d’intrusion (source: https://www.ionos.fr/digitalguide/serveur/securite/systemes-de-detection-et-de-prevention-dintrusion/)

### a) Systèmes de détection d'intrusion réseaux (NIDS)

Les systèmes de détection d'intrusion réseaux (ou NIDS : Network Intrusion Detection System) sont les IDS les plus répandus. Ce sont des outils très utiles pour l'administrateur réseaux qui va pouvoir, en temps réel, comprendre ce qui se passe sur son réseau et prendre des décisions en ayant toutes les informations.

Ils peuvent être placés à divers endroits sur le réseau, en amont ou en aval d'un pare feu ou encore sur chaque hôte, comme un anti virus.


### b) Systèmes de détection d'intrusion hôtes (HIDS)

Les systèmes de détection d'intrusion hôte (ou HIDS : Host-based Intrusion Detection System) sont des IDS mis en place directement sur les hôtes à surveiller. Ils vont directement analyser les fichiers de l'hôte, les différents appels système et aussi les événements réseaux. Par conséquent ces analyses sont strictement limitées à l'hôte sur laquelle l'HIDS est installé et n'ont aucune vue sur le réseau.

Les HIDS agissent comme des antivirus mais en plus poussé, car les antivirus ne sont intéressés que par les activités malveillantes du poste alors qu'un HIDS va pouvoir intervenir s'il détecte des attaques par dépassement de tampon et concernant les processus système par exemple39.

Les features de notre dataset peuvent diviser en  quatre catégories.: Intrinsic, Content, Host-based, and Time-based. Voici une description de ces différentes catégories.

Intrinsic features: peut être dérivé de l'en-tête du paquet sans examiner la charge utile elle-même et contenir les informations de base sur le paquet. Cette catégorie contient les features 1 à 9.

Content features : contiennent des informations sur les paquets originaux, car ils sont envoyés en plusieurs morceaux plutôt qu'un seul. Avec ces informations, le système peut accéder à la charge utile. Cette catégorie contient les fonctionnalités 10 à 22.

Time-based : analysent le trafic entrant sur une fenêtre de deux secondes et contiennent des informations telles que le nombre de connexions qu'il a tenté d'établir avec le même hôte. Ces caractéristiques sont principalement des décomptes et des taux plutôt que des informations sur le contenu du trafic entrant. Cette catégorie contient les fonctionnalités 23 à 31.


Host-based features : Sont similaires aux features Time-Base, sauf qu'au lieu d'analyser sur une fenêtre de 2 secondes, il analyse sur une série de connexions établies (combien de requêtes adressées au même hôte sur un nombre x de connexions). Ces fonctionnalités sont conçues pour accéder aux attaques qui durent plus d’une période de temps de deux secondes. Cette catégorie contient les fonctionnalités 32 à 41.

4 Categorical (Features: 2, 3, 4, 42)
6 Binary (Features: 7, 12, 14, 20, 21, 22)
23 Discrete (Features: 8, 9, 15, 23–41, 43)
10 Continuous (Features: 1, 5, 6, 10, 11, 13, 16, 17, 18, 19)

```python
from sklearn.preprocessing import LabelEncoder

def label_encode_categorical_columns(train_data, test_data):
    # Make a copy of the original DataFrames to preserve them
    train_encoded = train_data.copy()
    test_encoded = test_data.copy()
    label_encoders = {}

    # Initialize LabelEncoder
    encoder = LabelEncoder()

    # Fit and transform categorical columns in training data
    for col in train_data.select_dtypes(include=['object']).columns:
        train_encoded[col] = encoder.fit_transform(train_data[col])
        label_encoders[col] = dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))

        # Transform corresponding columns in test data
        if col in test_data.columns:
            test_encoded[col] = encoder.transform(test_data[col])
        else:
            # If a categorical column doesn't exist in test data, handle it accordingly
            print(f"Column '{col}' not found in test data. Skipping...")

    return train_encoded, test_encoded, label_encoders

# Usage
train_encoded_data, test_encoded_data, encoders_dict = label_encode_categorical_columns(train_data2, test_data2)

# Check if the encodings are consistent across train and test data
print("Train Encoders Dictionary:")
for col, mapping in encoders_dict.items():
    print(f"\n{col} Encoding:")
    print(mapping)
```

```python
dataset = pd.concat([train_encoded_data, test_encoded_data], ignore_index=True)
train_dataNormalized=train_encoded_data.copy()
test_dataNormalized=test_encoded_data.copy()
dataset_Normalized=dataset.copy()
def normalize(df, cols):
    result = df.copy() # do not touch the original df
    for feature_name in cols:
        max_value = dataset[feature_name].max()
        min_value = dataset[feature_name].min()
        if max_value > min_value:
            result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result

#Normalizing training set
train_dataNormalized = normalize(train_dataNormalized,train_encoded_data.columns[:41])
test_dataNormalized = normalize(test_dataNormalized,test_encoded_data.columns[:41])
dataset_Normalized = normalize(dataset_Normalized,test_encoded_data.columns[:41])
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Assuming models is a list containing your models (knn_model, svm_model, clf_model, rf_model, lr_model)
models = [knn_model, svm_model, clf_model, rf_model, lr_model]
X_trainF, X_testF, y_trainF, y_testF = train_test_split(dataX, dataY, test_size=0.2, random_state=42)

# Assuming X_test is your test data and y_test is the corresponding true labels
for class_index in range(5):  # Loop through each class
    plt.figure(figsize=(8, 6))
    plt.title(f'ROC Curve - Class {class_index}')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')

    for model in models:  # Loop through each model
        # Assuming the models have predict_proba method for probability predictions
        if hasattr(model, "predict_proba"):
            y_prob = model.predict_proba(X_testF)
        else:  # For models that don't have predict_proba
            y_prob = model.decision_function(X_testF)
            y_prob = (y_prob - y_prob.min(axis=0)) / (y_prob.max(axis=0) - y_prob.min(axis=0))

        # Get the probability of the positive class for the current class_index
        if isinstance(y_prob, np.ndarray):
            prob_pos = y_prob[:, class_index]
        else:
            prob_pos = y_prob[class_index]


        fpr, tpr, _ = roc_curve((y_testF == class_index).astype(int), prob_pos)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f'{type(model).__name__} (AUC = {roc_auc:.2f})')

    plt.plot([0, 1], [0, 1], 'k--', label='Random')
    plt.legend(loc='lower right')
    plt.show()
```
